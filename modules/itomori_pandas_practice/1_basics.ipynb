## Pandas Basics

```python
import pandas as pd
```

`Dataframe` is the core data structure in pandas. They are basically just tables.

```python
carrot_df = pd.DataFrame(
    {
        "CarrotId": [1, 2, 3, 4, 5],
        "Color": ["#FFCC00", "#FF8B00", "#D25F2A", "#F0B347", "#F38C22"],
        "Length": [5, 4.3, 4.7, 6.7, 6],
        "Price": [1, 0.8, 0.9, 1.3, 0.9],
    }
)

carrot_df
```
