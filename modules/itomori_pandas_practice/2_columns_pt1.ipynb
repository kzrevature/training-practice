## Columns (pt 1)

The `Column` abstraction is essential for analyzing and manipulating data.

```python
import pandas as pd

grocery_df = pd.read_csv("data/grocery_catalog.csv")

# use the column name as an index
grocery_df["name"]
```

Many basic operators are overloaded for Columns.

```python
# arithmetic
grocery_df["price"] + 1000
```

```python
# dot product
grocery_df["price"] * grocery_df["quantity"]
```

```python
# str concatenation
"prefix_" + grocery_df["name"]
```

```python
# length (number of rows)
len(grocery_df["calories"])

# note that this also works on the df as a whole
# len(grocery_df)
```

There are also 'normal' functions.

```python
# map()
grocery_df["price"].map(lambda x: x**2)

# apply() works too
# grocery_df["price"].apply(lambda x: x ** 2)
```

```python
# str.contains()
grocery_df["name"].str.contains("a")
```

```python
# unique()
grocery_df["price"].unique()
```

```python
# sort_values()
grocery_df["name"].sort_values()
```

```python
# sum()
grocery_df["quantity"].sum()
```
