## Columns (pt 2)

How can we manipulate columns in context of an entire DataFrame?

```python
import pandas as pd

grocery_df = pd.read_csv("data/grocery_catalog.csv")
```

Adding columns to a DataFrame.

```python
grocery_df["total_price"] = grocery_df["quantity"] * grocery_df["price"]
grocery_df["10x_quantity"] = grocery_df["quantity"] * 10
grocery_df["ratings"] = [4.8, 4.0, 3.4, 4.5, 3.9, 4.3, 2.9]

grocery_df
```

```python
# without mutating the existing DataFrame
grocery_df_new = grocery_df.assign(new_col=[0, 0, 0, 0, 0, 0, 0])
len(grocery_df.columns), len(grocery_df_new.columns)
```

Selecting specific columns from a DataFrame.

```python
grocery_df[["id", "name", "price"]]
```

```python
# you can also make a new DataFrame from these isolated columns
grocery_df_narrowed = grocery_df[["id", "name", "price"]]
```

Removing columns from a DataFrame

```python
# this doesn't mutate the original DataFrame
grocery_df.drop(columns=["10x_quantity"])
```
