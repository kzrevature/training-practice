```python
import pandas as pd

grocery_df = pd.read_csv("data/grocery_catalog.csv")
```

Filtering based on a condition.

```python
grocery_df[grocery_df["category"] != "Fruit"]

# notice how this evaluates to:
# grocery_df[[False, False, True, True, True, False, True]]
```
